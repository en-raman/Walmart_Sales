```python
df['unit_price']=df['unit_price'].str.replace('$','').astype(float)
```

```python
df.to_csv("Walmart_Cleaned.csv", index=False)
```

```python
# show help for the DataFrame.to_sql method without calling it
help(df.to_sql)
```

```python
# create a SQLAlchemy engine to connect to the PostgreSQL database
engine = create_engine('postgresql://postgres:nopass@localhost:5432/Walmart_db')
try:
    # write the DataFrame to a SQL table named 'sales_data'
    df.to_sql('sales_data', engine, if_exists='replace', index=False)
    print("DataFrame successfully written to the database.")
except Exception as e:
    print(f"An error occurred: {e}")
```
